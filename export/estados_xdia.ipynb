```python
import pandas as pd
import matplotlib.pyplot as plt
```

## lectura parquet tramites totales del procedimiento

```python
df_tramites_proc_X = pd.read_parquet('../data/tramita/tramites_autoconsumo.parquet')
```

```python
columns_estados = ['id_exp','desc_tramite','fecha_tramite']
df_estados_proc_X = df_tramites_proc_X[columns_estados]
```

### filtro columnas id, fecha, estado

```python
df_estados_proc_X = df_estados_proc_X.assign(fecha=df_estados_proc_X['fecha_tramite'].dt.date)
```

```python
df_estados_proc_X.head()
```

### limpieza procedimientos que acaban en presenacion de solicitud

```python
final_states = df_estados_proc_X.sort_values(['id_exp', 'fecha_tramite']).groupby('id_exp').last().reset_index()
final_states
```

```python
final_states[final_states['id_exp']==1084489]
```

```python
final_states_grp = final_states.groupby('desc_tramite').size()
final_states_grp
```

```python
id_exp_to_remove = final_states.loc[(final_states['desc_tramite'] =='1-1-Presentación de la solicitud'), "id_exp"].unique()
id_exp_to_remove
```

```python
len(id_exp_to_remove)
```

```python
# Filter out rows with those `id_exp` values
filtered_df = df_estados_proc_X[~df_estados_proc_X["id_exp"].isin(id_exp_to_remove)]
```

```python
len(filtered_df.id_exp.unique())
```

## ordenado y pivotaje

```python
# tabla con el número total de cambios de estado , cada día, una columna por estado
filtered_df = filtered_df.sort_values(by='fecha_tramite')
# no hay valores para todas las fechas, habría que rellenar las fechas vacías.
df_pivot_estados = filtered_df.pivot_table(index='fecha', columns='desc_tramite', aggfunc='size', fill_value=0)
```

```python
df_pivot_estados.head()
```

```python
df_pivot_estados.to_parquet('../data/tramita/cambiosestado_xdia_autoconsumo.parquet')
```

```python
df_pivot_estados.tail()
```

## plot

```python
df_pivot_estados.columns
```

```python
columns_to_plot = ['1-1-Presentación de la solicitud', 'Registro de solicitud','42-1-Propuesta Resolución Estimatoria de Concesión']  # Specify the columns you want to plot

df_pivot_estados[columns_to_plot].plot(figsize=(12, 6), title="Cambios de estado por día")
plt.xlabel("Fecha")
plt.ylabel("Númeo de cambios de estado")
plt.grid(True)
plt.show()
```
