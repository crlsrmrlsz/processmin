```python
import pandas as pd
```

```python
df_def_procedimientos= pd.read_parquet('./data/tramita/def_procedimientos.parquet')
df_def_procedimientos.head()
```

```python
unique_combinations_ori = df_def_procedimientos[['NPROCADMI','NUMTRAM_ORI', 'DENOMINACION_ORIGEN']].drop_duplicates()
unique_combinations_ori_unif = unique_combinations_ori.rename(columns={'NPROCADMI': 'cod_proc', 'NUMTRAM_ORI': 'num_tramite', 'DENOMINACION_ORIGEN' : 'denom' })
unique_combinations_des = df_def_procedimientos[['NPROCADMI','NUMTRAM_DES', 'DENOMINACION_DESTINO']].drop_duplicates()
unique_combinations_des_unif = unique_combinations_des.rename(columns={'NPROCADMI': 'cod_proc', 'NUMTRAM_DES': 'num_tramite', 'DENOMINACION_DESTINO' : 'denom' })

unique_combinations = pd.concat([unique_combinations_ori_unif, unique_combinations_des_unif], ignore_index=True).drop_duplicates()
unique_combinations
```

```python
unique_combinations.to_parquet('./data/tramita/def_proc_tramite_unicos.parquet')
```
