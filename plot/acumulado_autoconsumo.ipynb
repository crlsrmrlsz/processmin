## posibles estados finales extraidos del dataframe

```python
df_tramites = pd.read_parquet('../data/tramita/tramites_autoconsumo.parquet')

#filtro columnas
columns_estados = ['id_exp','desc_tramite','fecha_tramite']
df_tramites = df_tramites[columns_estados]

#ordenado
df_tramites = df_tramites.sort_values(by=['id_exp', 'fecha_tramite'])

# Create a new column for the next state in the process
df_tramites['next_state'] = df_tramites.groupby('id_exp')['desc_tramite'].shift(-1)
df_tramites['fecha_fin'] = df_tramites.groupby('id_exp')['fecha_tramite'].shift(-1)

# Drop rows where the next state is NaN (final states)
transitions = df_tramites.dropna(subset=['next_state'])[['desc_tramite', 'next_state', 'fecha_tramite','fecha_fin']]

# Group by current state and get all unique next states
following_states = transitions.groupby('desc_tramite')['next_state'].unique().to_dict()

from pprint import pprint
pprint(following_states)
```
