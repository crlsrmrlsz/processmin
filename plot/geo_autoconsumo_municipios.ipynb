```python
import json

# Load the GeoJSON file
with open("../data/geo/cnig_otros/delimit_municipios/Municipios.geojson", "r", encoding="utf-8") as f:
    geojson_data = json.load(f)

# Add an 'id' to each feature based on 'properties.name' (or any other field)
for feature in geojson_data["features"]:
    feature["id"] = feature["properties"]["CODIGOINE"]

# Save the modified GeoJSON if needed
with open("../data/geo/cnig_otros/delimit_municipios/Municipios_id_codine.geojson", "w", encoding="utf-8") as f:
    json.dump(geojson_data, f)
```

## agregamos datos por municipios


```python
df_agg = df_autoconsumo_simple.groupby(['codine', 'municipio']).agg(
    total_processes=('id_exp', 'count'),
    online_processes=('es_telematica', 'sum'),
    person_processes=('dni', lambda x: x.notnull().sum())
).reset_index()
df_agg.head()
```

```python
# Add calculated percentages
total_processes = df_agg['total_processes'].sum()
df_agg['percent_of_total'] = (df_agg['total_processes'] / total_processes) * 100
df_agg['percent_online'] = (df_agg['online_processes'] / df_agg['total_processes']) * 100
df_agg['percent_person'] = (df_agg['person_processes'] / df_agg['total_processes']) * 100
df_agg.head()
```
