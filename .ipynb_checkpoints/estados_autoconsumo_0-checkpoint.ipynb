```python

import pandas as pd
import numpy as np

# Sample input dataframe
data = {
    'id_exp': [1, 1, 2, 2],
    'num_tramite': [1, 2, 1, 3],
    'fecha': ['2023-01-01', '2023-01-03', '2023-01-02', '2023-01-04']
}
df = pd.DataFrame(data)
df['fecha'] = pd.to_datetime(df['fecha'])

# Set MultiIndex
df = df.set_index(['id_exp', 'fecha']).sort_index()

# Generate timeline for reindexing
timeline = pd.date_range(start=df.index.get_level_values('fecha').min(),
                         end=df.index.get_level_values('fecha').max())

# Forward-fill approach
df_filled = df.unstack('num_tramite').reindex(
    pd.MultiIndex.from_product([df.index.levels[0], timeline]), method='ffill'
)

# Flatten MultiIndex columns
df_filled = df_filled.stack().reset_index(name='num_tramite')

# Aggregate to count processes per state per date
result_ffill = df_filled.groupby(['fecha', 'num_tramite']).size().unstack(fill_value=0)

print("Forward-Fill Approach Result:")
print(result_ffill)



```

```python
# For-loop approach
df = df.reset_index()  # Ensure the dataframe isn't MultiIndexed for this approach
df = df.sort_values(by=['id_exp', 'fecha'])

result = []
for _, group in df.groupby('id_exp'):
    states = group['num_tramite'].values
    dates = group['fecha'].values
    for i in range(len(states)):
        start_date = dates[i]
        end_date = dates[i + 1] - pd.Timedelta(days=1) if i + 1 < len(dates) else timeline[-1]
        date_range = pd.date_range(start=start_date, end=end_date)
        result.append(pd.DataFrame({'fecha': date_range, 'num_tramite': states[i]}))

expanded_df = pd.concat(result)

# Count processes by state for each date
result_loops = expanded_df.groupby(['fecha', 'num_tramite']).size().unstack(fill_value=0)

print("For Loop Approach Result:")
print(result_loops)
```

```python
import time

# Generate a larger dataset for testing
n_processes = 1000
data = {
    'id_exp': np.repeat(np.arange(1, n_processes + 1), 3),
    'num_tramite': np.tile([1, 2, 3], n_processes),
    'fecha': pd.to_datetime(np.tile(['2023-01-01', '2023-01-05', '2023-01-10'], n_processes))
}
df_large = pd.DataFrame(data).set_index(['id_exp', 'fecha']).sort_index()

# Measure Forward-Fill Approach
start_time = time.time()
df_large_filled = df_large.unstack('num_tramite').reindex(
    pd.MultiIndex.from_product([df_large.index.levels[0], pd.date_range('2023-01-01', '2023-01-10')]),
    method='ffill'
)
df_large_filled = df_large_filled.stack().reset_index(name='num_tramite')
result_ffill_large = df_large_filled.groupby(['fecha', 'num_tramite']).size().unstack(fill_value=0)
ffill_time = time.time() - start_time

# Measure For Loop Approach
start_time = time.time()
df_large_reset = df_large.reset_index()
result_loop_large = []
for _, group in df_large_reset.groupby('id_exp'):
    states = group['num_tramite'].values
    dates = group['fecha'].values
    for i in range(len(states)):
        start_date = dates[i]
        end_date = dates[i + 1] - pd.Timedelta(days=1) if i + 1 < len(dates) else pd.Timestamp('2023-01-10')
        date_range = pd.date_range(start=start_date, end=end_date)
        result_loop_large.append(pd.DataFrame({'fecha': date_range, 'num_tramite': states[i]}))
expanded_df_large = pd.concat(result_loop_large)
result_loops_large = expanded_df_large.groupby(['fecha', 'num_tramite']).size().unstack(fill_value=0)
loop_time = time.time() - start_time

# Print Results
print(f"Forward-Fill Approach Time: {ffill_time:.4f} seconds")
print(f"For Loop Approach Time: {loop_time:.4f} seconds")
```
